```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img = cv2.imread('buildings.jpg', 1)
```

```python
img.shape
```

```python
left = cv2.filter2D(img, -1, np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]]))
right = cv2.filter2D(img, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
top = cv2.filter2D(img, -1, np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]]))
bottom = cv2.filter2D(img, -1, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
A = np.hstack([np.vstack([top, bottom]), np.vstack([left, right])])
A.shape
```

```python
img = cv2.imread('sails.png')
img.shape
left = cv2.filter2D(img, -1, np.array([[3, 0, -3], [10, 0, -10], [3, 0, -3]]))
right = cv2.filter2D(img, -1, np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]))
top = cv2.filter2D(img, -1, np.array([[3, 10, 3], [0, 0, 0], [-3, -10, -3]]))
bottom = cv2.filter2D(img, -1, np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]))
plt.imshow(left)
plt.show()
plt.imshow(right)
plt.show()
plt.imshow(top)
plt.show()
plt.imshow(bottom)
plt.show()
```

```python
img2 = cv2.filter2D(img, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
plt.figure(figsize=(16, 10.8))
plt.imshow(img2)
```
