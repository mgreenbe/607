```python
from __future__ import absolute_import, division, print_function
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Dense, Flatten
from tensorflow.keras.utils import to_categorical
import numpy as np
from matplotlib import pyplot as plt

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_04 = X_train[y_train <= 4].reshape(-1, 28**2)
y_train_04 = to_categorical(y_train[y_train <= 4])
X_test_04 = X_test[y_test <= 4, :].reshape(-1, 28**2)
y_test_04 = to_categorical(y_test[y_test <= 4])
```

```python
I = np.random.choice(len(X_train_04), size=500, replace=False)
U = X_train_04[I]
v = y_train_04[I]
```

```python
num_classes = 5

model = Sequential([
    Dense(num_classes, activation='softmax', input_dim=28**2, name="output_layer")
])


model.compile(loss='categorical_crossentropy',
              optimizer='adadelta',
              metrics=['accuracy'])

model.fit(U, v, epochs=1000, validation_data=(X_test_04, y_test_04))
```

```python
plt.plot(range(1000), model.history.history["val_accuracy"])
```

```python
from sklearn.linear_model import LogisticRegression
R = LogisticRegression()
R.fit(U, np.argmax(v, axis=1))
y_p = R.predict(X_test_04)
yy = np.argmax(y_test_04, axis=1)
```
